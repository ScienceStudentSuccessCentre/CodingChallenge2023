# Modification 1: Colour Space Convertion

As you may have noticed, OpenCV reads the images in blue colour instead of the characteristic Carleton red. This is because, by default, OpenCV reads the image as Blue, Green, Red (BGR) instead of Red, Blue, Green (RGB). Therefore, you have to change the colour space used to read the image. A colour space describes a specific, measurable, and fixed range of possible colours and luminance values" [[1](https://blog.frame.io/2020/02/03/color-spaces-101/#What_is_color_space)].

The following function transforms one colour space to another, with the colour space being denoted by a cv2-provided constant [[2](https://docs.opencv.org/3.4/d8/d01/group__imgproc__color__conversions.html#gga4e0972be5de079fed4e3a10e24ef5ef0a4f17670b4ee8da649dc368629230c990)].

`cv1.cvtColor(source, color_space)`

- source: image read using cv2
- color_space: ColorConvertionCode

```python
photoA = cv.cvtColor(photoA, cv.COLOR_BGR2RGB)
photoB = cv.cvtColor(photoB, cv.COLOR_BGR2RGB)
```

```python
# Run this code to see the images
plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.title("Photo A")
plt.imshow(photoA)
plt.subplot(1, 2, 2)
plt.title("Photo B")
plt.imshow(photoB)
```
